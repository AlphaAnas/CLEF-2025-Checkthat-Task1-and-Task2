## Prepare Data

```python
train_data = pd.read_csv("/content/train-spa.csv") # change path and file name
val_data = pd.read_csv("/content/dev-spa.csv")

train_data = train_data.sample(frac=1).reset_index(drop=True)
val_data = val_data.sample(frac=1).reset_index(drop=True)

ds = DatasetDict({
        'train': Dataset.from_pandas(train_data),
        'validation': Dataset.from_pandas(val_data)
        })
```

```python
def tokenize_sample_data(data):
    # Max token size is set to 1024 and 128 for inputs and labels, respectively.
    input_feature = tokenizer(data["post"], truncation=True, max_length=1024)
    label = tokenizer(data["normalized claim"], truncation=True, max_length=128)
    return {
        "input_ids": input_feature["input_ids"],
        "attention_mask": input_feature["attention_mask"],
        "labels": label["input_ids"],
    }


tokenized_ds = ds.map(
    tokenize_sample_data,
    remove_columns=["post", "normalized claim"],
    batched=True,
    batch_size=1)


def tokenize_sentence(arg):
    encoded_arg = tokenizer(arg)
    return tokenizer.convert_ids_to_tokens(encoded_arg.input_ids)
```

```python
# Install required packages
!pip install -q transformers rouge_score
!pip install -q bert_score  # Optional for BERTScore

# Import necessary libraries
import numpy as np
import nltk
from nltk.translate import meteor_score
from evaluate import load
import re

# Download required NLTK data
nltk.download('punkt')
nltk.download('wordnet')

# Load metrics from HuggingFace evaluate
rouge = load('rouge')
```

```python
def compute_metrics(eval_pred):
    predictions, labels = eval_pred

    # Decode generated summaries
    decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens=True)

    # Replace -100 in the labels with pad token id
    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    # Clean up decoded texts
    decoded_preds = [pred.strip() for pred in decoded_preds]
    decoded_labels = [label.strip() for label in decoded_labels]

    # Compute ROUGE scores (primary metric for summarization)
    rouge_result = rouge.compute(
        predictions=decoded_preds,
        references=decoded_labels,
        use_stemmer=False,  # Better for Spanish, which has different morphology than English
        tokenizer=lambda x: x.split()  # Simple Spanish tokenization
    )

    # Calculate METEOR score (handles synonyms and stems well for Spanish)
    meteor_scores = []
    for pred, ref in zip(decoded_preds, decoded_labels):
        # For Spanish, we use simple whitespace tokenization
        pred_tokens = pred.lower().split()
        ref_tokens = ref.lower().split()
        score = meteor_score.meteor_score([ref_tokens], pred_tokens)
        meteor_scores.append(score)

    # Calculate BERTScore (semantic similarity, works well cross-lingually)
    try:
        from bert_score import score
        P, R, F1 = score(decoded_preds, decoded_labels, lang="es", verbose=False)
        bertscore = F1.mean().item()
    except ImportError:
        bertscore = 0.0

    # Compute basic statistics
    prediction_lens = [len(pred.split()) for pred in decoded_preds]
    reference_lens = [len(label.split()) for label in decoded_labels]

    # Compile metrics (no multiplication by 100)
    result = {key: value for key, value in rouge_result.items()}
    result["meteor"] = np.mean(meteor_scores)
    result["bertscore"] = bertscore

    # Add length statistics
    result["gen_len"] = np.mean(prediction_lens)
    result["ref_len"] = np.mean(reference_lens)
    result["compression_ratio"] = np.mean([p/r if r > 0 else 0 for p, r in zip(prediction_lens, reference_lens)])

    return {k: round(v, 4) for k, v in result.items()}
```

```python
training_args = Seq2SeqTrainingArguments(
    output_dir = "./results/bart-post-simplifier-20epochs",
    eval_strategy="epoch",
    num_train_epochs = 20,
    learning_rate = 5e-5,           # Reduced from 3e-4 for more stable training
    lr_scheduler_type = "cosine",   # Often better than linear for convergence
    warmup_ratio = 0.1,             # Using ratio instead of steps (10% of training)
    optim = "adamw_torch",
    weight_decay = 0.01,
    per_device_train_batch_size = 4,
    per_device_eval_batch_size = 4,
    gradient_accumulation_steps = 16,
    predict_with_generate = True,
    generation_max_length = 128,
    logging_steps = 10,             # This is fine for visibility
    save_strategy = "epoch",        # Save checkpoints each epoch
    load_best_model_at_end = True,  # Load best model after training
    metric_for_best_model = "eval_loss", # Or another metric you compute
    push_to_hub = False,
    fp16 = True                     # Mixed precision training if GPU supports it
)
trainer = Seq2SeqTrainer(
    model = model,
    args = training_args,
    data_collator = data_collator,
    compute_metrics = compute_metrics,
    train_dataset = tokenized_ds["train"],
    eval_dataset = tokenized_ds["validation"],
    tokenizer = tokenizer
)
```

```python
def simplify_test_posts(test_df, model, tokenizer):
    """Generate simplifications for test posts without reference claims"""
    results = []

    # Get the device that the model is on
    device = model.device

    for i, row in tqdm(test_df.iterrows(), total=len(test_df), desc="Generating simplifications"):
        # Get input text
        input_text = row['post']

        # Tokenize the input
        inputs = tokenizer(input_text, return_tensors="pt", max_length=512, truncation=True, padding="max_length")

        # Move input tensors to the same device as the model
        inputs = {k: v.to(device) for k, v in inputs.items()}

        # Generate simplified text
        with torch.no_grad():  # Add this to prevent unnecessary gradient computation
            output_ids = model.generate(
                inputs["input_ids"],
                max_length=128,
                num_beams=4,
                early_stopping=True,
                no_repeat_ngram_size=2
            )

        # Decode the output
        simplified = tokenizer.decode(output_ids[0], skip_special_tokens=True)

        results.append({
            "original_post": input_text,
            "simplified_claim": simplified
        })

    # Create a DataFrame with results
    results_df = pd.DataFrame(results)

    # Save predictions to CSV
    results_df.to_csv('test_predictions_kor.csv', index=False)

    print(f"Generated simplifications for {len(results_df)} test examples")
    print("Saved predictions to 'test_predictions.csv'")

    # Display a few examples
    print("\nSample predictions:")
    for i, row in results_df.head(3).iterrows():
        print(f"\nOriginal: {row['original_post']}")
        print(f"Simplified: {row['simplified_claim']}")

    return results_df

# Generate predictions for test data
test_predictions = simplify_test_posts(test_df, model, tokenizer)
```

```python
def create_simplification_pipeline(model_path="./results/bart-post-simplifier-20epochs"):
    """Create a pipeline for simplifying social media posts"""
    from transformers import BartForConditionalGeneration, BartTokenizer

    # Load the model and tokenizer
    model = BartForConditionalGeneration.from_pretrained(model_path)
    tokenizer = BartTokenizer.from_pretrained(model_path)

    def simplify(post):
        # Tokenize the input
        inputs = tokenizer(post, return_tensors="pt", max_length=512, truncation=True, padding="max_length")

        # Generate simplified text
        output_ids = model.generate(
            inputs["input_ids"],
            max_length=128,
            num_beams=4,
            early_stopping=True,
            no_repeat_ngram_size=2
        )

        # Decode the output
        simplified = tokenizer.decode(output_ids[0], skip_special_tokens=True)
        return simplified

    return simplify

# Create the pipeline
simplify = create_simplification_pipeline()

# Example usage on a few complex posts
# Example usage on a few complex posts
sample_posts = [
  "La compleja interacción de factores socioeconómicos y decisiones políticas ha precipitado una transformación sin precedentes en las dinámicas del mercado, lo que exige estrategias adaptativas en diversos sectores empresariales. 💼📉🌍",
  "Las arquitecturas de computación cuántica que aprovechan los fenómenos de entrelazamiento demuestran ventajas potenciales en la reducción de la complejidad algorítmica para dominios problemáticos que antes eran intratables con los paradigmas de la computación clásica. 🧠⚛️💻",
  "¡Es un milagro! 😱😱 El gobierno prometió arreglar las calles... otra vez. 😂😂 #DéjàVu #Elecciones"
]

print("\nTesting the simplification pipeline on sample posts:")
for post in sample_posts:
    simplified = simplify(post)
    print(f"\nOriginal: {post}")
    print(f"Simplified: {simplified}")
```
